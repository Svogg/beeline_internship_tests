## 6. Вывести агрегированную инфорацию по количеству книг в диапазонах:
## 0 - 1
## 1 - 2
## 2 - 3
## 3 - 4
## 4 - 5

```python
df.insert(0, 'Количество книг', [el for el in range(1, len(df) + 1)], False)
```

```python
df['average_rating'] = pd.cut(df['average_rating'], bins=[0, 1, 2, 3, 4, 5], right=True, include_lowest=True)
```

```python
df.groupby([df['average_rating']])['Количество книг'].agg(['describe'])
```
